```python
cars_significant_variables = data[['price', 'fueltype', 'aspiration','carbody', 'drivewheel','wheelbase',
                  'curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'boreratio','horsepower', 
                    'fueleconomy', 'carlength','carwidth', 'carsrange']]
cars_significant_variables.head()
```

```python
sns.pairplot(cars_significant_variables)
```

```python
def dummies(x,df):
    temp = pd.get_dummies(df[x], drop_first = True)
    df = pd.concat([df, temp], axis = 1)
    df.drop([x], axis = 1, inplace = True)
    return df
# Applying the function to the cars_significant_variables

cars_significant_variables = dummies('fueltype',cars_significant_variables)
cars_significant_variables = dummies('aspiration',cars_significant_variables)
cars_significant_variables = dummies('carbody',cars_significant_variables)
cars_significant_variables = dummies('drivewheel',cars_significant_variables)
cars_significant_variables = dummies('enginetype',cars_significant_variables)
cars_significant_variables = dummies('cylindernumber',cars_significant_variables)
cars_significant_variables = dummies('carsrange',cars_significant_variables)
```

### Train-Test Split and feature scaling¶

```python
from sklearn.model_selection import train_test_split

data_train, data_test = train_test_split(cars_significant_variables, train_size = 0.7, test_size = 0.3, random_state = 100)
```

_Transforms features by scaling each feature to a given range._

    X_std = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))<br>
    X_scaled = X_std * (max - min) + min

where min, max = feature_range.

The transformation is calculated as:

    X_scaled = scale * X + min - X.min(axis=0) * scale
    where scale = (max - min) / (X.max(axis=0) - X.min(axis=0))



```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
num_vars = ['wheelbase', 'curbweight', 'enginesize', 'boreratio', 'horsepower','fueleconomy','carlength','carwidth','price']
data_train[num_vars] = scaler.fit_transform(data_train[num_vars])
```

_Highly correlated variables to price are - curbweight, enginesize, horsepower,carwidth and highend._

```python
y_train = data_train.pop('price')
X_train = data_train
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm 
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

_the goal of recursive feature elimination (RFE) is to select features by recursively considering smaller and smaller sets of features. First, the estimator is trained on the initial set of features and the importance of each feature is obtained either through a coef_ attribute or through a feature_importances_ attribute._

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
rfe = RFE(lm , 10)
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
X_train.columns[rfe.support_]
```

```python
X_train_rfe = X_train[X_train.columns[rfe.support_]]
X_train_rfe.head()
```

```python
def build_model(X,y):
    X = sm.add_constant(X) #Adding the constant
    lm = sm.OLS(y,X).fit() # fitting the model
    print(lm.summary()) # model summary
    return X
def checkVIF(X):
    vif = pd.DataFrame()
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return(vif)
```

```python
X_train_new = build_model(X_train_rfe,y_train)
```

__p-vale of twelve seems to be higher than the significance value of 0.05, hence dropping it as it is insignificant in presence of other variables.__

```python
X_train_new = X_train_rfe.drop(["twelve"], axis = 1)
X_train_new = build_model(X_train_new,y_train)
```

```python
X_train_new = X_train_new.drop(["fueleconomy"], axis = 1)
X_train_new = build_model(X_train_new,y_train)
```

```python
checkVIF(X_train_new)
```

```python
X_train_new = X_train_new.drop(["curbweight"], axis = 1)
```

```python
X_train_new = build_model(X_train_new,y_train)
```

```python
checkVIF(X_train_new)
```

```python
X_train_new = X_train_new.drop(["sedan"], axis = 1)
X_train_new = build_model(X_train_new,y_train)
```

```python
checkVIF(X_train_new)
```

```python
X_train_new = X_train_new.drop(["wagon"], axis = 1)
X_train_new = build_model(X_train_new,y_train)
```

```python
checkVIF(X_train_new)
X_train_new = X_train_new.drop(["dohcv"], axis = 1)
X_train_new = build_model(X_train_new,y_train)
checkVIF(X_train_new)
```

```python
lm = sm.OLS(y_train,X_train_new).fit()
y_train_price = lm.predict(X_train_new)
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_price), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)   
```

__Error terms seem to be approximately normally distributed, so the assumption on the linear modeling seems to be fulfilled.__
